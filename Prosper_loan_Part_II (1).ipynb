```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_loan= pd.read_csv("prosperLoanData.csv")
```

## (Visualization 1)

This first visualisation shows the occupation of borrowers and it can be seen clearly that people with other occcupation not listed are likely to request for loans. Also professionals and computer programmers have high potential of reqesting for loans


```python
# lets create a function that labels our plot
def chart_labels(x_label,y_label,title):
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
```

```python
plt.figure(figsize = [10, 5]) 
base_color = sb.color_palette()[0]
sb.countplot(data = df_loan, x = 'Occupation', color = base_color,
             order = df_loan['Occupation'].value_counts().head(10).index );
chart_labels('Occupation','Counts', 'Borrowers Occupation')
plt.xticks(rotation = 90)
plt.show()
```

## (Visualization 2)

This visualisation shows that occupation of borrowers and it can be seen clearly that people with other occcupation not listed are and th employed individuals likely to request for loans. Also professionals and computer programmers have high potential of reqesting for loans

```python
plt.figure(figsize = [10, 5]) 
base_color = sb.color_palette()[0]
sb.countplot(data = df_loan, y = 'EmploymentStatus', color = base_color)
chart_labels('Employment Status','Counts', 'Employment Status of all Borrowers"')
plt.show()
```

## (Visualization 3)

This visalisation shows the top 10 borrowing state in the US. This visualisation shows that Carlifornia is the state that has individuals that borros most followed bt Texas. We Can categorically say that we have more borroin in these states as a results of their population

```python
plt.figure(figsize = [10, 5]) 
base_color = sb.color_palette()[0]
sb.countplot(data = df_loan, x= 'BorrowerState', color = base_color, 
             order = df_loan['BorrowerState'].value_counts().head(10).index)
chart_labels('Borrower State','Counts', 'Top 10 borrowing state ')
plt.xticks(rotation = 45)
plt.show()
```
